```python
import psutil

# Set thresholds (in percentages)
MEMORY_THRESHOLD = 80  # Warning if more than 80% of RAM is used
DISK_THRESHOLD = 90    # Warning if more than 90% of Disk space is used

# Monitor memory usage
def monitor_memory():
    memory = psutil.virtual_memory()
    print("=== Memory Usage ===")
    print(f"Used memory: {memory.used / 1024**3:.2f} GB")
    print(f"Available memory: {memory.available / 1024**3:.2f} GB")
    print(f"Total memory: {memory.total / 1024**3:.2f} GB")
    print(f"Memory usage: {memory.percent}%")

    # Check if memory usage exceeds threshold
    if memory.percent > MEMORY_THRESHOLD:
        print(f"⚠️ Warning: Memory usage exceeded {MEMORY_THRESHOLD}%!")
    print("\n")

# Monitor disk space usage
def monitor_disk():
    disk = psutil.disk_usage('/')
    print("=== Disk Space Usage ===")
    print(f"Used disk space: {disk.used / 1024**3:.2f} GB")
    print(f"Free disk space: {disk.free / 1024**3:.2f} GB")
    print(f"Total disk space: {disk.total / 1024**3:.2f} GB")
    print(f"Disk usage: {disk.percent}%")

    # Check if disk usage exceeds threshold
    if disk.percent > DISK_THRESHOLD:
        print(f"⚠️ Warning: Disk usage exceeded {DISK_THRESHOLD}%!")
    print("\n")

# Run both monitors
monitor_memory()
monitor_disk()
```

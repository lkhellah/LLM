```python
import psutil

# Set thresholds (in percentages)
MEMORY_THRESHOLD = 80  # Warning if more than 80% of RAM is used
DISK_THRESHOLD = 90    # Warning if more than 90% of Disk space is used

# Monitor memory usage
def monitor_memory():
    memory = psutil.virtual_memory()
    print("=== Memory Usage ===")
    print(f"Used memory: {memory.used / 1024**3:.2f} GB")
    print(f"Available memory: {memory.available / 1024**3:.2f} GB")
    print(f"Total memory: {memory.total / 1024**3:.2f} GB")
    print(f"Memory usage: {memory.percent}%")

    # Check if memory usage exceeds threshold
    if memory.percent > MEMORY_THRESHOLD:
        print(f"⚠️ Warning: Memory usage exceeded {MEMORY_THRESHOLD}%!")
    print("\n")

# Monitor disk space usage
def monitor_disk():
    disk = psutil.disk_usage('/')
    print("=== Disk Space Usage ===")
    print(f"Used disk space: {disk.used / 1024**3:.2f} GB")
    print(f"Free disk space: {disk.free / 1024**3:.2f} GB")
    print(f"Total disk space: {disk.total / 1024**3:.2f} GB")
    print(f"Disk usage: {disk.percent}%")

    # Check if disk usage exceeds threshold
    if disk.percent > DISK_THRESHOLD:
        print(f"⚠️ Warning: Disk usage exceeded {DISK_THRESHOLD}%!")
    print("\n")

# Run both monitors
monitor_memory()
monitor_disk()
```

# Model benchmark

```python
messages = []
for prompt in [entry["Question"] for entry in dataset]:
  messages.append([
      {"role": "system", "content": "Please reason step by step, and put your final answer within \\boxed{}. I'll ask multiple questions at once, separate them by blank lines"},
      {"role": "user", "content": prompt}
  ])
```

```python
monitor_memory()
monitor_disk()
```

```python
import time
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "Qwen/Qwen2.5-Math-1.5B-Instruct"
device = "cuda" # the device to load the model onto

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

start_time = time.time()

def chat_with_qwen(prompt):
  text = tokenizer.apply_chat_template(
      prompt,
      tokenize=False,
      add_generation_prompt=True
  )


  model_inputs = tokenizer([text], return_tensors="pt").to(device)

  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=512000
  )
  generated_ids = [
      output_ids[len(input_ids):] for input_ids, output_ids in zip(model_inputs.input_ids, generated_ids)
  ]

  return tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]

responses = []
for message in messages:
    response = chat_with_qwen(message)
    responses.append(response)

print(f"Time taken: {time.time() - start_time}")
```

```python
for index in range(len(responses)):
  print(f"Question {index + 1}:", dataset[index]["Question"])
  print(responses[index], "\n\n\n")
```
